```python
#Name: Qurratul Ayen Elma
#ID: 20201121
#SECTION: 11

import math
from queue import PriorityQueue

heuristic_values = {}
graph = {}

with open('/Users/chowdhuryelma/Downloads/input.txt', mode='r') as input_file:
    lines = input_file.read().splitlines()

for i in lines: #each line
    parts = i.split() #name val
    city = parts[0] #Name
    heuristic = int(parts[1]) #Val
    heuristic_values[city] = heuristic
    #print(heuristic_values)

    connections = []
    for j in range(2, len(parts)-1, 2):
        neighboring_node = parts[j]
        distance = int(parts[j + 1])
        connections.append((neighboring_node, distance)) #(Zerind, 75), (Timisoara, 118), and (Sibiu, 140)

    graph[city] = connections # Arad: (Zerind, 75), (Timisoara, 118), and (Sibiu, 140)
    #print(graph[city])


# heuristic_values = {
#     "Arad": 366,
#     "Bucharest": 0,
#     "Craiova": 160,
#     "Dobreta": 242,
#     "Eforie": 161,
#     "Fagaras": 176,
#     "Giurgiu": 77,
#     "Hirsova": 151,
#     "Iasi": 226,
#     "Lugoj": 244,
#     "Mehadia": 241,
#     "Neamt": 234,
#     "Oradea": 380,
#     "Pitesti": 100,
#     "RimnicuVilcea": 193,
#     "Sibiu": 253,
#     "Timisoara": 329,
#     "Urziceni": 80,
#     "Vaslui": 199,
#     "Zerind": 374,}
# graph = {
      # "Arad":
      # ("Zerind", 75),
      # ("Timisoara", 118),
      # ("Sibiu", 140),
      # "Craiova":
      # ("Dobreta", 120),
      # ("RimnicuVilcea", 146),
      # ("Pitesti", 138),
      # "Eforie":
      # ("Hirsova", 86),
      # "Fagaras":
      # ("Sibiu", 99),
      # ("Bucharest", 211),
      # "Giurgiu":
      # ("Bucharest", 90),
      # "Mehadia":
      # ("Lugoj", 70),
      # ("Dobreta", 75),
      # "Neamt":
      # ("Iasi", 87),
      # "Sibiu":
      # ("Oradea", 151),
      # ("Arad", 140),
      # ("RimnicuVilcea", 80),
      # ("Fagaras", 99),
      # "Oradea":
      # ("Zerind", 71),
      # ("Sibiu", 151),
      # ("Arad", 140),
      # "Pitesti":
      # ("RimnicuVilcea", 97),
      # ("Craiova", 138),
      # ("Bucharest", 101),
      # "RimnicuVilcea":
      # ("Sibiu", 80),
      # ("Craiova", 146),
      # ("Pitesti", 97),
      # "Dobreta":
      # ("Mehadia", 75),
      # ("Craiova", 120),
      # "Hirsova":
      # ("Urziceni", 98),
      # ("Eforie", 86),
      # "Iasi":
      # ("Vaslui", 92),
      # ("Neamt", 87),
      # "Lugoj":
      # ("Timisoara", 111),
      # ("Mehadia", 70),
      # ("Timisoara", 111),
      # "Urziceni":
      # ("Bucharest", 85),
      # ("Hirsova", 98),
      # ("Vaslui", 142),
      # ...
      # }


#-----------------------------------------------------------------------------
# A* search function
def a_star_search(graph, start, goal):
    priority_queue = [(0, start)]  # Priority queue, (f(n), node)
    #print('priorty_queue')
    #print(priority_queue)
    came_from_parent = {}  # Parent_Node for back track
    #print('came_from_parent')
    #print(came_from_parent)
    g_n = {city: math.inf for city in heuristic_values} #all cities -> +ve infinity
    #print('g_n')
    #print(g_n)
    g_n[start] = 0 #Arad: 0

    while priority_queue:
        _, current = min(priority_queue)
        #print('_, current')
        #print(_, current) #f(n), City
        priority_queue.remove((_, current))
        #print('priority_queue')
        #print(priority_queue)

        if current == goal: #(Arad != Bucharest)
          #print(current) #Arad
          #print(goal) #Bucharest
          path = [] #optimal path -> back track
          while current:
              path.insert(0, current)
              #print('path')
              #print(path)
              current = came_from_parent.get(current)
              #print('current')
              #print(current)

          return path, g_n[goal]

        for neighboring_node, distance in graph.get(current, []):
          #print('neighboring_node:', neighboring_node)
          #print('distance:',distance)
          #print('g_n[current]:',g_n[current])
          tentative_g_n = g_n[current] + distance
          #print('tentative_g_n',tentative_g_n)
          #print('g_n[neighboring_node]:',g_n[neighboring_node])
          if tentative_g_n < g_n[neighboring_node]:
                #print('came_from_parent[neighboring_node]',came_from_parent[neighboring_node])
                came_from_parent[neighboring_node] = current
                #print('came_from_parent[neighboring_node]',came_from_parent[neighboring_node])
                g_n[neighboring_node] = tentative_g_n
                #print('g_n[neighboring_node]',g_n[neighboring_node])
                #print('tentative_g_n',tentative_g_n)
                #print('heuristic_values[neighboring_node]',heuristic_values[neighboring_node])
                f_n = tentative_g_n + heuristic_values[neighboring_node]
                #print('f(n)', f_n)
                #print('neighboring_node',neighboring_node)
                priority_queue.append((f_n, neighboring_node))
                #print('priority_queue',priority_queue)


    return "NO PATH FOUND", 0 #no path found from the Start node to the End node

#---------------------------------------------------------------------------------
start_node = input("Start Node: ") #Arad
destination = input("Destination: ") #Bucharest

# path and cost
path, cost = a_star_search(graph, start_node, destination)
#path, cost = a_star_search(graph,Arad,Bucharest)
if cost != 0:
    print("Path:", " -> ".join(path)) #arad->Sibiu
    print("Total distance:"+str(cost)+" km")
else:
    print(path)
```
