```python
import numpy as np
import math
import random

# (2) Fitness Func
#================================================================

def fitness(individual, transactions):
    total = 0
    lgth = len(individual)
    #print(lgth)
    for i in range(lgth):
      #print(i)
      #print(individual[i])
      if individual[i] == 1: # if true = 1
        a = transactions[i]
        #print(a)
        total = total + a
        #print(total)

    return abs(total)

# (3) Crossover
#================================================================

def crossover(prnt_1, prnt_2):

  length = len(prnt_1)
  #print(length)
  till = length - 1
  #print(till)
  crossover_point = random.randint(1, till)
  #print(crossover_point)

  c1_part1 = prnt_1[:crossover_point]
  #print(c1_part1)
  c1_part2 = prnt_2[crossover_point:]
  #print(c1_part2)

  child_1 = c1_part1 + c1_part2
  #print(child_1)

  c2_part1 = prnt_2[:crossover_point]
  #print(c2_part1)
  c2_part2 = prnt_1[crossover_point:]
  #print(c2_part2)

  child_2 = c2_part1 + c2_part2
  #print(child_2)

  return child_1, child_2

#(4) Mutation
#================================================================

def mutate(individual, mutation_rate=0.01):
    mutated_individual = []
    for bit in individual:
      #print(bit)
      if random.random() < mutation_rate:
        flip=int(not bit)
        #print(flip)
        mutated_individual.append(flip)
        #print(mutated_individual)
      else:
        mutated_individual.append(bit)
        #print(mutated_individual)
    return mutated_individual

# (5) Population Generation
#================================================================

def create_population(pop_size, num_transactions):
    initial_population = []
    #print(pop_size)
    #print(num_transactions)
    for i in range(pop_size):
     # print(i)
      individual = []
      for j in range(num_transactions):
        #print(j)
        e = random.choice([1, 0]) #e is gene==True(1)->include transaction & gene==False(0)->exclude transaction
        #print(e)
        individual.append(e)
        #print(individual)
      initial_population.append(individual)
      #print(initial_population)
    return initial_population

#================================================================

def get_fitness_score(individual_fitness_tuple):
  w = individual_fitness_tuple[1]
  return w

#MAIN FUNC
#================================================================

def genetic_algorithm(transactions, pop_size=100, max_generations=100000): #number of daily transactions,N=100 ; amount of transaction, S=100000
    population = create_population(pop_size, len(transactions))
    #print('ELMA')
    #print(population)
    for i in range(max_generations):
      #print(i) #generations
      population_fitness = []
      for individual in population:
        #print(individual)
        individual_fitness = fitness(individual, transactions)
        #print(individual_fitness)
        population_fitness.append((individual, individual_fitness))
        #print(population_fitness)
      for individual, individual_fitness in population_fitness:
        #print(individual, individual_fitness)
        if individual_fitness == 0:
          result = ""
          for gene in individual:
            #print(gene)
            result += str(gene)
            #print(result)
          return result

      population_sorted = sorted(population_fitness, key=get_fitness_score)
      #print(population_sorted)
      parents = []
      number_of_parents = pop_size // 2
      #print(number_of_parents)
      for i in range(number_of_parents):
        #print(i)
        x=0
        val=population_sorted[i][x]
        #print(val)
        parents.append(val)
        #print(parents)

      new_generation = []
      lngth = len(new_generation)
      while lngth < pop_size:
          prnt_1, prnt_2 = random.sample(parents, 2)
          #print(prnt_1)
          #print(prnt_2)
          child_1, child_2 = crossover(prnt_1, prnt_2)
          m1 = mutate(child_1)
          #print(m1)
          m2 = mutate(child_2)
          #print(m2)
          new_generation.extend([m1, m2])
          #print(new_generation)

      population = new_generation
      #print(population)

    return -1  # No solution found

#================================================================
#================================================================

input_file = 'input.txt'
with open(input_file, 'r') as file:
    lines = file.readlines()
    #print(lines)
    transactions = []
    for line in lines[1:]:
      #print(line)
      parts = line.split()
      #print(parts)
      amount = int(parts[1])
      #print(amount)
      if parts[0] == 'l':
        multiplier = -1
      else:
        multiplier = 1
      trc = amount * multiplier
      #print(trc)
      transactions.append(trc)
      #print(transactions)

result = genetic_algorithm(transactions)
print(result)
```
